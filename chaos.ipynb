```python
#Rossler System
```

```python
def next_rossler(x0, y0, z0, dt):
    a=b=0.2
    c=5.7
    dxdt = -y0-z0
    dydt = x0+a*y0
    dzdt = b+z0*(x0-c)
    dx=dxdt*dt
    dy=dydt*dt
    dz=dzdt*dt
    x=x0+dx
    y=y0+dy
    z=z0+dz
    return x, y, z
    #print'x,y,z = %s.' % x,y,z
    


    
```

```python
ros1 = []
ros2 = []
ros3 = []
x=0
y=0
z=0
for i in range(30000):
    x1,y1,z1=next_rossler(x,y,z,0.01)
    #print'%s.' %x1,y1,z1
    x=x1
    y=y1
    z=z1
    ros1.append(x)
    ros2.append(y)
    ros3.append(z)
%matplotlib inline

from mpl_toolkits import mplot3d
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
plt.figure()
ax = plt.axes(projection='3d')
#plt.plot(out2,out3)
ax.plot3D(ros1, ros2, ros3, 'b')
```

```python
#Lorenz system
```

```python
def next_lorenz(x0, y0, z0, dt):
    sig=10
    b=8/3
    r=28
    dxdt = sig*(y0 - x0)
    dydt = r*x0 -y0- x0*z0
    dzdt = x0*y0-b*z0
    dx=dxdt*dt
    dy=dydt*dt
    dz=dzdt*dt
    x=x0+dx
    y=y0+dy
    z=z0+dz
    return x, y, z
```

```python
lor1 = []
lor2 = []
lor3 = []
x=15
y=20 
z=30
for i in range(5000):
    x1,y1,z1=next_lorenz(x,y,z,0.01)
    #print'%s.' %x1,y1,z1
    x=x1
    y=y1
    z=z1
    lor1.append(x)
    lor2.append(y)
    lor3.append(z)
%matplotlib inline

from mpl_toolkits import mplot3d
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
plt.figure()
#plt.subplot(311)
ax = plt.axes(projection='3d')
#plt.plot(out1,'mediumblue')

#plt.subplot(312)
#plt.plot(out2,'mediumblue')

#plt.subplot(313)
#plt.plot(out3,'mediumblue')
plt.xlabel('Samples')
ax.plot3D(lor1, lor2, lor3, 'g')
```

```python
#Chen system
```

```python
def next_chen(x0, y0, z0, dt):
    a, b, c = 40, 3, 28
    dxdt = -a*(x0 - y0)
    dydt = (c-a)*x0 +c*y0- x0*z0
    dzdt = -b*z0+x0*y0
    dx=dxdt*dt
    dy=dydt*dt
    dz=dzdt*dt
    x=x0+dx
    y=y0+dy
    z=z0+dz
    return x, y, z
```

```python
che1 = []
che2 = []
che3 = []
x,y, z = -0.1, 0.5, -0.6
for i in range(10000):
    x1,y1,z1=next_chen(x,y,z,0.001)
    #print'%s.' %x1,y1,z1
    x=x1
    y=y1
    z=z1
    che1.append(x)
    che2.append(y)
    che3.append(z)
%matplotlib inline

from mpl_toolkits import mplot3d
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
plt.figure()
ax = plt.axes(projection='3d')
#plt.plot(out2,out3)
ax.plot3D(che1, che2, che3, 'y')
```

```python
def next_chua(x0, y0, z0, dt):
    a, b, m0, m1 = 15.6, 28, -1.143, -0.714
    #f0 = m1*x0+0.5*(m0-m1)*(abs(x0+1)-abs(x0-1))
    dxdt = a*(y0 - x0- (m1*x0+0.5*(m0-m1)*(abs(x0+1)-abs(x0-1))))
    dydt = x0-y0+z0
    dzdt = -b*y0
    dx=dxdt*dt
    dy=dydt*dt
    dz=dzdt*dt
    x=x0+dx
    y=y0+dy
    z=z0+dz
    return x, y, z
```

```python
chu1 = []
chu2 = []
chu3 = []
x,y, z = 0.7, 0.0, 0.0
for i in range(3000):
    x1,y1,z1=next_chua(x,y,z,0.01)
    #print'%s.' %x1,y1,z1
    x=x1
    y=y1
    z=z1
    chu1.append(x)
    chu2.append(y)
    chu3.append(z)
%matplotlib inline

from mpl_toolkits import mplot3d
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
plt.figure()
ax = plt.axes(projection='3d')
#plt.plot(out2,out3)
ax.plot3D(chu1, chu2, chu3, 'k')
```
